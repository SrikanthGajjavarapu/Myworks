```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
df = pd.read_csv('Testdata.csv',low_memory=False)
df
```

```python
df = df[['tradeId', 'version', 'clDateTime',
       'cDateTime',  'rate',
       'expirationDate', 'Timestamp', 'quantity',
       'endDate', 'price',
       'PartyId', 'transactionType', 'Reporting Status']].copy()
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
```

```python
df.isna().sum()
```

```python
df.loc[~df.duplicated(subset = ['version'])]
```

```python
ax = df['version'].value_counts().plot(kind='bar', title = 'total versions')
ax.set_xlabel('no of versions')
ax.set_ylabel('count')
```

```python
df['rate'].plot(kind='hist')
```

```python
df_corr = df[['tradeId','version','rate','price','PartyId']].dropna().corr()
```

```python
sns.heatmap(df_corr, annot=True)
```

```python
df.query('version').groupby('version')['tradeId'].agg(['mean','count']).sort_values('mean')['mean'].plot(kind='bar')
```
